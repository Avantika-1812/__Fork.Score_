```python
import re
import pandas as pd
from nltk.stem import PorterStemmer

# Load your dataset (adjust the path as needed)
dataset = pd.read_csv("C:\\Users\\Avantika Singh\\Downloads\\reviews1.csv")

# Initialize the corpus and the Porter Stemmer
corpus = []
ps = PorterStemmer()

# Loop through the dataset
for i in range(len(dataset)):
    # Check if the review is not NaN
    if pd.notna(dataset['review'][i]):
        # Cleaning the review text
        zomato_ratings = re.sub('[^a-zA-Z]', ' ', dataset['review'][i])
        zomato_ratings = zomato_ratings.lower()
        zomato_ratings = zomato_ratings.split()
        
        # Stemming the words
        zomato_ratings = [ps.stem(word) for word in zomato_ratings]
        
        # Joining the stemmed words back into a single string
        zomato_ratings = ' '.join(zomato_ratings)
        
        # Appending the cleaned review to the corpus
        corpus.append(zomato_ratings)

# Now 'corpus' contains the cleaned and stemmed reviews
print(corpus)
```

```python
#Bags of word creation
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the CountVectorizer with the desired max_features
cv = CountVectorizer(max_features=1600)

# Fit and transform the corpus into a feature array
X = cv.fit_transform(corpus).toarray()

# Assuming `y` is the last column in your dataset (such as a label or rating)
y = dataset.iloc[:, -1].values
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler

# 1. Load the dataset
file_path = 'C:\\Users\\Avantika Singh\\Downloads\\zomato_reviews.csv'  

data = pd.read_csv(file_path)

# 2. Data Exploration
print(data.info())  # Check data types, null values
print(data.describe())  # Statistical summary of numeric columns
print(data.head())  # First few rows of the dataset

# 3. Handling Duplication of Data
data = data.drop_duplicates()
print("Duplicates removed, if any.")

# 4. Handling Missing Values
# Option 1: Drop rows with any missing values
data = data.dropna()

# Option 2: Impute missing values for numerical columns
# data['column_name'] = data['column_name'].fillna(data['column_name'].mean())

# Check for missing values after handling
print(data.isnull().sum())

# 5. Handling Categorical Data
# Identify categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns

# Label Encoding for ordinal categorical columns (replace with appropriate columns)
label_encoder = LabelEncoder()
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col])

# One-Hot Encoding for nominal categorical columns (use drop_first=True to avoid dummy trap)
data = pd.get_dummies(data, drop_first=True)

# 6. Splitting the Dataset into Training and Testing Sets
X = data.drop('review', axis=1) 
y = data['rating']  

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 7. Feature Scaling
# Standardization
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Alternatively, use Normalization
# normalizer = MinMaxScaler()
# X_train = normalizer.fit_transform(X_train)
# X_test = normalizer.transform(X_test)

print("Data preprocessing complete.")
```
